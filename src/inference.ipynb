```python
import pickle
import torch
```

```python
from word_vector import WordVector
```

```python
pretrained_wv = WordVector('latest', 'skipgram')
```

```python
with open('../data/validation.txt', 'r') as f:
    validation = f.readlines()
for i in range(len(validation)):
    validation[i] = validation[i].strip().split()
```

```python
def solve_analogy(wv, pair, query):
    try:
        first_word, second_word = pair
        result_vector = wv[query] - wv[first_word] + wv[second_word]
        word, scores = wv.find_k_similar_from_vector(result_vector, k=4)
        return word, scores
    except KeyError as e:
        print(f'KeyError on {e}')
        return "", None
```

```python
correct = 0
for val in validation:
    try:
        first, second, query, true_label = val
    except:
        continue
    word, scores = solve_analogy(pretrained_wv, (first.lower(), second.lower()), query.lower())
    try:
        print(f'{first.lower()}: {second.lower()}; {query.lower()}: {list(scores.keys())[1]}')
    except:
        continue
    if true_label in list(scores.keys()):
        correct += 1
```
